## 2. Missing Values
Tujuan : Mencari missing value dan membuat keputusan untuk missing value.

```python
missing_value = df.isnull().sum()
print("Data yang kosong : ", missing_value)
```

Insight : Tidak ada data yang kosong, semua terisi dengan baik.

Keputusan : Tidak perlu membuat keputusan untuk mengatasi missing value.

## 3. Outliers
Tujuan : Menemukan dan mengatasi outlier (data dengan nilai yang menyimpang jauh) pada data.

```python
import seaborn as sns
import matplotlib.pyplot as plt

numeric_features = ["tenure", "MonthlyCharges"]
# SeniorCitizen masih bersifat boolean karena hanya ada 0 dan 1.

for feature in numeric_features:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=df[feature])
    plt.title(f'Box Plot of {feature}')
    plt.show()
```

Insight : Tidak ditemukan outlier pada data.

Keputusan: Tidak perlu mengambil keputusan.
